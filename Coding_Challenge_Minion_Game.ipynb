###  4. Implement the solution and test it using example inputs. Fix bugs, if any.

The solution follows the description of the game best illustrated by the image. A small difference is that instead of iterating over the string for consonents first and then for vowels as per the game instructions, there is a single iteration across the string with the sub-strings identified for vowels and consonants being separated during that iteration.

```python
# Solution will need counter from collections

from collections import Counter
```

```python
def minion(str):   

    # Create variables for counters
    u1 = []
    u2 = []
    u1count = 0
    u2count = 0

    # Define what constitutes a vowel.  All else is a consonant.
    vowels ="AEIOU"

    # Iterate across string searching for sub-strings and creating a list for vowels (Stuart) and consonant (Kevin)
    for i in range(len(str)):
        if str[i] in vowels:
            for k in range(len(str[i:])):
                u1.append(str[i:i+k+1])

        else:
            for l in range(len(str[i:])):
                u2.append(str[i:i+l+1])

    # Count up by grouping sub-strings withni the two lists            
    u1 = Counter(u1)
    u2 = Counter(u2)

    # Sum up the number of sub-strings for Stuart and for Kevin
    for i in u1.values():
        u1count = u1count + i
    for i in u2.values():
        u2count = u2count + i
 
    # Finally return whichever person has won with their score, or return a draw with the score.
    if u1count > u2count:        
        return('Kevin',u1count)   
    elif u1count < u2count:        
        return('Stuart',u2count)   
    else:   
        return('Draw',u1count)
```

Come with the optimized correct solution and explain it in simple words below:

1. Work from the first letter to the last letter in the string and determine whether it is a consonant or a vowel.
2. The number of substrings available will always be the length from the letter to the end of the string.
3. Simply add the number of substrings to a running total for each of the two players, adding the number to Stuart if it is a consonant and Kevin if it is a vowel.

### 8. Implement the solution and test it using example inputs. Fix bugs, if any.

```python
def minion_optimised(string):
    
    # Create variables for counters
    vowels_list ="AEIOU"
    consonants = 0
    vowels = 0

    # Work through the string calculating mathematically the number of sub-strings and adding them to the running totals
    n = len(string)
    for i, l in enumerate(string):
        if l in vowels_list:
            vowels += n-i
        else:
            consonants += n-i
            
    # Finally return whichever person has won with their score, or return a draw with the score.
    if vowels > consonants:        
        return('Kevin',vowels)   
    elif vowels < consonants:        
        return('Stuart',consonants)   
    else:   
        return('Draw',vowels)
```
